```python
from abc import ABC,abstractmethod     #abc-Abstract base class-inbuilt python module
                                       #ABC- base class for defining abstract classes
                                       #@abstractmethod- decorator to mark method as abstact
```

```python
#Abstarction + Encapsulation
#----------------------------]

class BankAccount(ABC):               #Abstract class, cannot create object of BankAccount directly
    def __init__(self,name,balance):
        self.name=name
        self.__balance=balance  #encapsulated(private)

    @abstractmethod   #Abstact Method, every subclass must implement this method
    def calculate_interest(self):
        pass              #pass do nothing, here only define method name not logic

    def deposite(self, amount):
        self.__balance +=amount
        print(f"{amount} deposited successfully!")
        
    def withdraw(self,amount):
        if amount<=self.__balance:
            self.__balance -=amount
            print(f"{amount} withdraw successfully!")
        else:
            print("Insufficient Balance")

    def get_balance(self):
        return self.__balance
```

```python
#Inheritance + Polymorphism
#----------------------------
class SavingAccount(BankAccount):
    def calculate_interest(self):
        interest=self.get_balance()*0.05      #interest=5%
        print(f"Interest (Savings): {interest:.2f}")
        return interest

class CurrentAccount(BankAccount):
    def calculate_interest(self):
        interest=self.get_balance()*0.02      # 2% interest
        print(f"Interest (current): {interest:.2f}")
```

```python
#Main Excution
#-------------------------------------------------
def main():
    print("Welcome To Python Bank")
    acc_type=input("Enter Account Type(savings/current):").lower()
    name=input("Enter your name:")
    balance=float(input("Enter initial deposite"))
    if acc_type=="savings":
        account=SavingAccount(name,balance)           #object creation for saving account class
    else:
        account=CurrentAccount(name,balance)         #object cration for current account class

    while True:
        print("\n------Menu------")
        print("1.Deposite")
        print("2.Withdraw")
        print("3.Check Balance")
        print("4.Calculate Interest")
        print("5.Exit")

        choice=input("Enter your choice:")
        if choice=="1":
            amt=float(input("Enter amount:"))
            account.deposite(amt)
        elif choice=="2":
            amt=float(input("Enter amount:"))
            account.withdraw(amt)
        elif choice=="3":
            print(f"Current Balance:{account.get_balance()}")
        elif choice=="4":
            account.calculate_interest()
        elif choice=="5":
            print("Thank you for Banking with us!!")
            break
        else:
            print("Invalid option!!")
            
if __name__=="__main__":           #run main function only when the file is executed directly not when imported elsewhere
    main()                         #__name__ -built in variable that stores the module name
```
