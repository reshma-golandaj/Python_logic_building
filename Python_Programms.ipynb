```python
#Merge the tools(HR)

def merge_the_tools(string, k):
    l1=len(string)
    for i in range(0,l1,k):
        substring=string[i:i+k]
        #print(substring)
        unique_char=""
        for ch in substring:
            if ch not in unique_char:
                unique_char+=ch
        print(unique_char)
    return unique_char
if __name__ == '__main__':
    merge_the_tools("aabbcddad",3)
```

```python
#Interview preparation problem hacker rank -count if avg is grater than--
def countResponseTimeRegressions(responseTimes):
    # Write your code here
    n=len(responseTimes)
    count=0
    total=responseTimes[0]
    for i in range(1,n):
        avg=total/i
        if(responseTimes[i]>avg):
            count+=1
        total+=responseTimes[i]
    return count
        

if __name__ == '__main__':
    responseTimes_count = int(input().strip())

    responseTimes = []

    for _ in range(responseTimes_count):
        responseTimes_item = int(input().strip())
        responseTimes.append(responseTimes_item)

    result = countResponseTimeRegressions(responseTimes)

    print(result)
```

```python
#without using inbuilt function
def swap_case(s):
    new_string=""    #Resukt will strore here
    for ch in s:
        if("A"<=ch<="Z"):    #if char is from capital A to Z 
            new_string+=chr(ord(ch)+32)   #if A occures then convert to 97 means add 32
        elif("a"<=ch<="z"):  #if char is lower a to z
            new_string+=chr(ord(ch)-32)   #if a occures then convert 97 to 65 by sunstracting 32
        else:                #if char is any special character then keep as it is
            new_string+=ch
    return new_string

if __name__=="__main__":
    string=input()
    result=swap_case(string)
    print(result)

#ord(ch)==>gives ASCII Value
#chr(value)==>converts number back to chatacter
#difference beetween uppercase and lowercase in ASCII=32
#"A" to "Z"== 65 to 90 
#"a" to "z"==>97 to 122'''
```

```python
#find a substring from a string nd give count of overlaping substring
def find_count_substring(string,sub_string):
    count=0
    start=0  #position where the search begins
    while True:
        idx=string.find(sub_string,start)  #find() will gives index position of substring where the substring first appears
        print(idx)
        if idx==-1:      #if idx is -1 then substring is not available in string
            break
        count+=1
        start=idx+1
    return count


if __name__=="__main__":
    string=input()
    sub_string=input()
    result=find_count_substring(string,sub_string)
    print(result)
```

```python
#Text wrap-given string divide into substring of given width--using loops

def wrap(string,max_width):
    lines=[]
    for i in range(0,len(string),max_width):
        lines.append(string[i:i+max_width])
    return "\n".join(lines)

if __name__=="__main__":
    string,max_width=input(),int(input())
    result=wrap(string,max_width)
    print(result)
```

```python
#Alphabetical Rangoli-Hacker rank
import string
size=int(input())
width=4*size-3
lines=[]    #empty list to store each row
alpha=string.ascii_lowercase   #it is simply a string containing a to z
for i in range(size):
    left=alpha[size-1:size-i-1:-1]   #descending letters
    right=alpha[size-i-1:size]          #ascending letters
    row="-".join(left+right)            #add "-" inbetween each letters
    lines.append(row.center(width,"-")) #crate each row by adding "-"
print("\n".join(lines+lines[-2::-1]))   #print full rangoli(mirror)
```
